```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/Yasser/Projects/masters_thesis/Master_Thesis/data/argument_relation_class.csv')
df.head()
```

```python
## I want to visualize the distribution of number of tokens of my data and draw a histogram

df['tokens'] = df['text'].apply(lambda x: len(x.split()))
df.head()

df['tokens'].describe() 


sns.set(style="darkgrid")
plt.figure(figsize=(12,6))
sns.histplot(df['tokens'], kde=False, color='red')
plt.title('Distribution of number of tokens', fontsize=20)
plt.xlabel('Number of tokens', fontsize=18)
plt.ylabel('Number of texts', fontsize=18)
plt.savefig('/Users/Yasser/Projects/masters_thesis/Master_Thesis/data/argument_relation_class.png')
plt.show()
```
